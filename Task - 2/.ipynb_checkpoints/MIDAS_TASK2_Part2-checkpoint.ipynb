# **MIDAS IIITD - Task 2 Part 2**

#### **Introduction**

For task two, the same dataset is used to train a new model, utilising only 0-9 training samples. Furthermore, this model is compared to new uninitialised model to compare performance on the MNIST dataset.

*   Contains 10 classes.
*   This includes handwritten digits 0-9.

This notebook only contains part two of the task. Other parts are in subsequent notebooks.

#### **Preprocessing**

1.   All images of the previous dataset are of shape (900,1200,3) by default.
2.   Their is no data imbalance.
3.   MNIST images are of shape (28,28,1) by default. 

Following steps are taken for preprocessing.

1. Reduce previous dataset images to 2 Dimensional image using Pillow. (900, 1200, 1)
2. Reduce size of image by rescaling to (128, 128).
3. Normalize image between 0-1 by dividing by 255.
4. Increase MNIST dataset images to (128,128).
5. Normalize in similar manner.

#### **Model and Training Details**

Two models are used as directed by the task:
1. Pretrained model on previous dataset - Retrained on MNIST training dataset.
2. Same network but randomly initialized.

**Pretrained Model**
This model is first trained on 0-9 image samples of Task 1 data. The model checkpoints are saved.
Then these checkoints are loaded back to the model and the MNIST data is fed to the model for fine tuning and retraining.
The final model is tested on MNIST test split.


**Random Initialised Network**
This model needs to be randomly initialised. On further study, it was seen that Keras by default uses **Glorot Initializer** (https://keras.io/api/layers/initializers/#glorotnormal-class).
For this implementation, instead of this, the **Random Normal Distribution** (https://keras.io/api/layers/initializers/#randomuniform-class) was chosen to initialize the weights for the network.

Categorical Cross Entropy is used as the loss function and Adam optimizer is used. Furthermore, ModelCheckpoints are added to :

  *   Save only the best model.
  *   Stop learning if the loss stagnates.
  *   Change learning rate of the model on the fly

# **Data Loading And Preprocessing**

#### **Task 1 Data**

```python
# Data stores the Numpy Array while Category is the class.

data = []
category = []

'''
Helper Function to:
1. Load all Images as Numpy Arrays.
2. Only 0-9 selected
'''

'''
Preprocessing:
1. Convert 3D image to 1D image to reduce computations.
2. Normalize between 0 and 1.
3. Keep data type as Float32.
4. Resize using pillow image library. Resize to (128,128) to improve speed.
'''

img_folder = 'train'

def load_data(img_folder):

    for dir1 in tqdm(os.listdir(img_folder)):
        
        # Select only 0-9
        dig = int(dir1[-3:])
        if (dig > 10):
            continue

        for file in os.listdir(os.path.join(img_folder, dir1)):
            image_path= os.path.join(img_folder, dir1,  file)
            pilload = Image.open(image_path).convert('L')
            pilload = pilload.resize((128,128))
            image= np.array(pilload)
            image = image.astype('float32')
            image /= 255  
            data.append(image)
            category.append(dir1)
            pilload.close()
```

```python
'''
Loading Dataset using function.
'''

load_data(img_folder)

'''
Create target dictionary
'''

target_dict = {k: v for v, k in enumerate(np.unique(category))}
target_val=  [target_dict[category[i]] for i in range(len(category))]
```

#### **MNIST Dataset**

```python
'''
Defining idx3 file paths
'''

train_images_path = 'MNIST/train-images.idx3-ubyte'
test_images_path = 'MNIST/t10k-images.idx3-ubyte'
train_labels_path = 'MNIST/train-labels.idx1-ubyte'
test_labels_path = 'MNIST/t10k-labels.idx1-ubyte'

train_data = idx2numpy.convert_from_file(train_images_path)
train_labels = idx2numpy.convert_from_file(train_labels_path)
test_data = idx2numpy.convert_from_file(test_images_path)
test_labels = idx2numpy.convert_from_file(test_labels_path)
```

```python
'''
Deining new function and method to resize all MNIST images to (128,128) and apply preprocessing methods
similar to those used for previous data.
'''
# New data initialisation
mnist_train = []
mnist_train_labels = []
mnist_test = []
mnist_test_labels = []

'''
Preprocessing:
1. Reload all images.
2. Normalize between 0 and 1.
3. Keep data type as Float32.
4. Resize using pillow image library. Resize to (128,128) to improve speed.
'''

def load_train_data(old_data, labels):

    for i in tqdm(range(0, len(old_data))):
        old = old_data[i]
        lab = labels[i]
        pilload = Image.fromarray(old.astype('uint8'))
        pilload = pilload.resize((128,128))
        image= np.array(pilload)
        image = image.astype('float32')
        image /= 255  
        mnist_train.append(image)
        mnist_train_labels.append(lab)
        pilload.close()


def load_test_data(old_data, labels):

    for i in tqdm(range(0, len(old_data))):
        old = old_data[i]
        lab = labels[i]
        pilload = Image.fromarray(old.astype('uint8'))
        pilload = pilload.resize((128,128))
        image= np.array(pilload)
        image = image.astype('float32')
        image /= 255  
        mnist_test.append(image)
        mnist_test_labels.append(lab)
        pilload.close()


load_train_data(train_data, train_labels)
load_test_data(test_data, test_labels)
```

## Data Setup

##### Data Augmentation is same as in part 1

```python
'''
Defining Training Data of Task1 Dataset
'''

X = np.array(data, np.float32).reshape(400,128,128,1)
Y = tf.keras.utils.to_categorical(target_val, 10)

del data
gc.collect()
```

```python
'''
Defining Data of MNIST
'''

X_mnist = np.array(mnist_train, np.float32).reshape(60000,128,128,1)
Y_mnist = tf.keras.utils.to_categorical(mnist_train_labels, 10)

del mnist_train
gc.collect()
```

## **Model Training**

### USING PRETRAINED MODEL

#### Training on Task 1 data

```python
'''
Defining Base Model and training data.
'''

model = keras.models.Sequential()

model.add(keras.layers.Conv2D(filters=12, kernel_size=(5,5), strides=2, activation='relu', 
                              input_shape=(128,128,1)))

model.add(keras.layers.Conv2D(filters=18, kernel_size=(3,3) , strides=2, activation='relu'))
model.add(keras.layers.Dropout(.5))

model.add(keras.layers.Conv2D(filters=24, kernel_size=(2,2), activation='relu'))

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(units=150, activation='relu'))
model.add(keras.layers.Dense(units=10, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
model.summary()
```

```python
'''
Setup DataGen
'''

datagen = ImageDataGenerator(
    featurewise_center=True, 
    featurewise_std_normalization=True,
    rotation_range = 10
)

datagen.fit(X)

train_generator = datagen.flow(X, Y, batch_size = 20)
```

```python
'''
Adding Callback API's to save best weights and change learning rate on the fly
Will be saved as task2_pretrained
'''
 
MCP = ModelCheckpoint('models/task2_pretrained.h5', verbose=1, save_best_only=True, monitor='accuracy', mode='max')
ES = EarlyStopping(monitor='accuracy', min_delta=0, verbose=0, restore_best_weights = True, patience=3, mode='max')
RLP = ReduceLROnPlateau(monitor='loss', patience=3, factor=0.2, min_lr=0.0001)
```

```python
%%time

'''
Training while also calculating time. 
THIS IS THE PRETRAINING STEP ON TASK 1 0-9 DATA
'''

tf.config.run_functions_eagerly(True)

history_pretraining = model.fit_generator(generator=train_generator,
                    epochs = 30, callbacks=[MCP,ES,RLP])
```

#### Using pretrained model on MNIST training.

```python
'''
Setup DataGen for MNIST
'''

datagen1 = ImageDataGenerator(
    featurewise_center=True, 
    featurewise_std_normalization=True,
    rotation_range = 10,
    validation_split = 0.2
)

datagen1.fit(X_mnist)

train_generator1 = datagen1.flow(X_mnist, Y_mnist, batch_size = 60, subset = 'training' )
val_generator1 = datagen1.flow(X_mnist, Y_mnist, batch_size = 60, subset = 'validation' )
```

```python
'''
Adding Callback API's to save best weights and change learning rate on the fly
Will be saved as task2_retrained
'''
 
MCP = ModelCheckpoint('models/task2_retrained.h5', verbose=1, save_best_only=True, monitor='val_accuracy', mode='max')
ES = EarlyStopping(monitor='val_accuracy', min_delta=0, verbose=0, restore_best_weights = True, patience=3, mode='max')
RLP = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.2, min_lr=0.0001)
```

```python
%%time

'''
ReTraining on MNIST while also calculating time. 
THIS IS THE RETRAINING STEP ON MNIST DATA. Hence, the previously trained pretrained model (task2_pretrained.h5) is loaded.
'''

# Change location, if needed, to location of task2_pretrained.h5 (Pretrained Model)
model = keras.models.load_model('models/task2_pretrained.h5')

tf.config.run_functions_eagerly(True)

history_mnist_retrain = model.fit_generator(generator=train_generator1,
                    validation_data=val_generator1,
                    epochs = 5, callbacks=[MCP,ES,RLP])
```

### USING RANDOM INIT MODEL

#### Training directly on MNIST

```python
'''
Defining Same Model. The model is kept same for comparision.
For each layer, the layer in initialised with random weights in normal distribution manner.
'''

model2 = keras.models.Sequential()

model2.add(keras.layers.Conv2D(filters=12, kernel_size=(5,5), strides=2, activation='relu', 
                              input_shape=(128,128,1)))

model2.add(keras.layers.Conv2D(filters=18, kernel_size=(3,3) , strides=2, activation='relu'))
model2.add(keras.layers.Dropout(.5))

model2.add(keras.layers.Conv2D(filters=24, kernel_size=(2,2), activation='relu'))

model2.add(keras.layers.Flatten())
model2.add(keras.layers.Dense(units=150, activation='relu'))
model2.add(keras.layers.Dense(units=10, activation='softmax'))

model2.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])
model2.summary()
```

```python
'''
Adding Callback API's to save best weights and change learning rate on the fly
Saved as task2_base.h5
'''

MCP = ModelCheckpoint('models/task2_base.h5', verbose=1, save_best_only=True, monitor='val_accuracy', mode='max')
ES = EarlyStopping(monitor='val_accuracy', min_delta=0, verbose=0, restore_best_weights = True, patience=3, mode='max')
RLP = ReduceLROnPlateau(monitor='val_loss', patience=3, factor=0.2, min_lr=0.0001)
```

```python
%%time

'''
Training on MNIST while also calculating time. 
We use the same training gen as before.
THIS IS THE RANDOMLY INIT MODEL TRAINED ON MNIST DATA
'''

# model1 = keras.models.load_model('models/task2_base.h5')

tf.config.run_functions_eagerly(True)

history_mnist = model2.fit_generator(generator=train_generator1,
                    validation_data=val_generator1,
                    epochs = 5, callbacks=[MCP,ES,RLP])
```

## **Testing Both Models on MNIST Test Split**

```python
'''
Setting Up MNIST Testing Data
'''

X_mnist_test = np.array(mnist_test, np.float32).reshape(10000,128,128,1)
Y_mnist_test = tf.keras.utils.to_categorical(mnist_test_labels, 10)
```

#### Testing Model 1 (Pretrained Model)

```python
# model = keras.models.load_model('models/task2_retrained.h5')


Pretrained_preds = model.predict(X_mnist_test)
scores = model.evaluate(X_mnist_test, Y_mnist_test)
print(scores)
```

```python
'''
Calculating Metrics for Model 1 (Pretrained)
'''

cm = confusion_matrix(np.argmax(Y_mnist_test, axis = 1), np.argmax(Pretrained_preds, axis = 1))
print(classification_report(np.argmax(Y_mnist_test, axis = 1), np.argmax(Pretrained_preds, axis = 1)))
```

```python
ax = sns.heatmap(cm, annot=True, fmt="d")
```

#### Testing Model 2 (Random Init)

```python
Random_preds = model2.predict(X_mnist_test)
scores = model2.evaluate(X_mnist_test, Y_mnist_test)
print(scores)
```

```python
'''
Calculating Metrics for Model 2 (Random Init)
'''

cm = confusion_matrix(np.argmax(Y_mnist_test, axis = 1), np.argmax(Random_preds, axis = 1))
print(classification_report(np.argmax(Y_mnist_test, axis = 1), np.argmax(Random_preds, axis = 1)))
```
